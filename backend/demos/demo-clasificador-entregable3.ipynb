# Demo Entregable 3: Clasificador de Narices

## Construyendo la máquina de clasificación

```python
clf = build_svm(cfg.params['SVM_POSITIVE'], cfg.params['SVM_NEGATIVE'])
```

## Cargando data a testear

```python
test_data = data_to_hist(cfg.params['SVM_TEST'])
```

## Generando las predicciones

```python
predictions = clf.predict(test_data)
d = {0:'invalida', 1:'valida'}
```

## Resultados

```python
for i in range(15):
    img = mpimg.imread(cfg.params['SVM_TEST']+str(i)+'.jpg', cv.IMREAD_COLOR)
    plt.axis('off')
    plt.title('Esta foto es clasificada {0}'.format(d[predictions[i]]))
    plt.imshow(img)
    plt.show()
```
