```python
%matplotlib inline
import numpy as np
import cv2 as cv
import config as cfg
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from train import build_svm, data_to_hist
```

## Construyendo la máquina de clasificación

```python
chi2_svm = build_svm(cfg.params['SVM_POSITIVE'], cfg.params['SVM_NEGATIVE'])
```

## Cargando data a testear

```python
test_data = data_to_hist(cfg.params['SVM_TEST'])
```

## Generando las predicciones

```python
predictions = chi2_svm.predict(test_data)
d = {0:'valida', 1:'invalida'}
```

## Resultados

```python
for i in range(10):
    img = mpimg.imread(cfg.params['SVM_TEST']+'00'+str(i)+'.jpg', cv.IMREAD_COLOR)
    plt.axis('off')
    plt.title('Esta foto es clasificada {0}'.format(d[predictions[i]]))
    plt.imshow(img)
    plt.show()
```
